```python
df = pd.read_excel(r"data/maliciousornot.xlsx")
```

```python
y = df["Type"].values
x = df.drop("Type", axis=1).values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=15)
```

```python
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = Sequential()

model.add(Dense(units=30, activation="relu"))
model.add(Dense(units=15, activation="relu"))
model.add(Dense(units=15, activation="relu"))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
early_stopping = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test, y_test), verbose=1, callbacks=[early_stopping,])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
model = Sequential()

model.add(Dense(units=30, activation="relu"))
model.add(Dropout(.6))

model.add(Dense(units=15, activation="relu"))
model.add(Dropout(.6))

model.add(Dense(units=15, activation="relu"))
model.add(Dropout(.6))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test, y_test), verbose=1, callbacks=[early_stopping,])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
predicts = model.predict(x_test) >= .5
predicts.astype(int)
```

```python
result = classification_report(y_test, predicts)
print(result)
```

```python
confusion_matrix(y_test, predicts)
```
